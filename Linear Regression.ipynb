### __Utilisation de la librairie Panda__

```python
# All imports
import pandas as pnd
from pandas import DataFrame
```

```python
data = pnd.read_csv("cost_revenue_clean.csv")

# Display the spreadsheet that we imported
data
```

```python
# Display the spreandsheet's infos
data.describe()
```

## __Utilisation du module dataframe__

```python
x = DataFrame(data, columns=["production_budget_usd"])
y = DataFrame(data, columns=["worldwide_gross_usd"])
```

## __Utilisation de la librairie Matplotlib__

```python
import matplotlib.pyplot as plt


plt.figure(figsize = (10,6))
plt.scatter(x,y, alpha=0.3)
plt.ylabel("Growth Revenue $")
plt.xlabel("Production costs $")
plt.title("Production Cost VS Growth Revenue")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

## __Utilisation de la librairie SCIKIT__

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(x,y)

regression.coef_
regression.intercept_

```

## __ReprÃ©sentation du graphe avec la rÃ©gression linÃ©aire__

```python

plt.figure(figsize = (10,6))
plt.scatter(x,y, alpha=0.3)

plt.plot(x,regression.predict(x), color = 'green', linewidth = 4)
plt.ylabel("Growth Revenue $")
plt.xlabel("Production costs $")
plt.title("Production Cost VS Growth Revenue")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()

```

## __Calcul du score obtenu par mon modÃ¨le__

```python
regression.score(x, y)
```
